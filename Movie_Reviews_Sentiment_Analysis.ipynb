```python
# Preprocessing: Rename columns and drop unnecessary ones
df.columns = ['review', 'sentiment'] + list(df.columns[2:])
df = df[['review', 'sentiment']]

# Encoding sentiment
df["Label"] = df["sentiment"].map({"positive":1,"negative":0})
```

```python
#remove number vs.
df['review'] = df['review'].str.lower().replace('[^\w\s]', '', regex=True).replace('\d+', '', regex=True).replace('\n', '', regex=True).replace('\r', '', regex=True)
df.head(5)
```

```python
# x y split
x=df["review"]
y=df["Label"]
```

```python
# vectorizing
vect=CountVectorizer(stop_words="english")
x=vect.fit_transform(x)
```

```python
# train test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
from sklearn.svm import LinearSVC

model = LinearSVC(max_iter=5000)
model.fit(x_train, y_train)
```

```python
#make predictions
predictions=model.predict(x_test)
```

```python
# model accuracy
accuracy_score(predictions,y_test)
```

```python
#save model
import joblib
joblib.dump(model,"model_svc.joblib")
```
